```python
import pandas as pd
import numpy as np
med = pd.read_csv("/content/KaggleV2-May-2016.csv")
house = pd.read_csv("/content/train.csv")
```

```python
missing_counts = house.isnull().sum()
print(missing_counts)
missing_counts_med = med.isnull().sum()
print(missing_counts_med)
```

```python
import matplotlib.pyplot as plt
missing_counts[missing_counts>0].plot(kind='bar',color = 'red')
plt.title('Count of Missing Values Per Column')
plt.ylabel('Number of Missing Values')
plt.show()
```

```python
to_plot = missing_counts_med[missing_counts_med > 0]

if not to_plot.empty:
    to_plot.plot(kind='bar', color='teal')
    plt.title("Missing Values in Medical Dataset")
    plt.show()
else:
    print("No missing values found! The Medical dataset is already clean.")
```

```python
def impute_data(df):
    num_cols = df.select_dtypes(include=['number']).columns
    for col in num_cols:
        df[col] = df[col].fillna(df[col].median())

    cat_cols = df.select_dtypes(include=['object']).columns
    for col in cat_cols:
        df[col] = df[col].fillna(df[col].mode()[0])

    return df

house_df = impute_data(house)
medical_df = impute_data(med)
```

```python
threshold = 0.40
limit = len(house_df) * threshold

house_df = house_df.dropna(thresh=limit, axis=1)
medical_df = medical_df.dropna(thresh=limit, axis=1)

print(f"Columns remaining after dropping high-missing features: {house_df.shape[1]}")
print(f"Columns remaining after dropping high-missing features: {medical_df.shape[1]}")
```

```python
print("Remaining missing values in House Prices:", house_df.isnull().sum().sum())
print("Remaining missing values in Medical Data:", medical_df.isnull().sum().sum())
```
